```python
from huggingface_hub import snapshot_download
dataset_path = snapshot_download(repo_id="RahulAloth/crf-dataset", repo_type="dataset")
dataset_path
   
```

```python
# 2. Choose your target folder
target_dir = os.path.expanduser("~/CRF/dataset/")
os.makedirs(target_dir, exist_ok=True)

# 3. Copy the dataset files
for root, dirs, files in os.walk(dataset_path):
    for filename in files:
        src_path = os.path.join(root, filename)
        dst_path = os.path.join(target_dir, filename)
        shutil.copy(src_path, dst_path)

print("All images are copied to:", target_dir)
```

```python
from ultralytics import YOLO
import cv2
import numpy as np
import os

# Load detection model
model = YOLO("/home/rahul/CRF/model/best.pt", task="detect")

# Load image
path = os.path.expanduser("~/CRF/dataset/crf_image_21.png")
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Run inference
# results = model(img)
results = model.predict(source=img, imgsz=1024, conf=0.5, half=False)

# Print detections
for r in results:
    for box in r.boxes:
        cls_id = int(box.cls)
        conf = float(box.conf)
        xyxy = box.xyxy[0].tolist()
        print(f"{model.names[cls_id]} | conf={conf:.2f} | box={xyxy}")
```
